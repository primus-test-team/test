```python
from random import randint
random = randint(1,1000)
life_count = 0
out_count = 0


while True:

    print("현재 out카운트:", out_count)
    if life_count == 9 or out_count == 3:
        YorN = input("계속하시겠습니까? (Y/N): ")
        if YorN == "Y":
            life_count = 0
            out_count = 0
            print("새로운 게임을 시작합니다")
            continue
        else:
            print("게임을 종료합니다")
            break

    custom_input = int(input("세자리 수 입력을 해주세요:"))

    first = custom_input // 100
    second = (custom_input // 10) % 10
    third = custom_input % 10

    print("첫번째 자리:", first)
    print("두번째 자리:", second)
    print("세번째 자리:", third)

    first_answer = random // 100
    second_answer = (random // 10) % 10
    third_answer = random % 10

    print("랜덤 숫자 첫번째 자리:", first_answer)
    print("랜덤 숫자 두번째 자리:", second_answer)
    print("랜덤 숫자 세번째 자리:", third_answer)

    #part1, strike
    if first == first_answer and second == second_answer and third == third_answer:
        print("홈런")
        break
    #part2, strike, ball, out
    elif first == first_answer or second == second_answer or third == third_answer:
        print("strike")
        life_count += 1

    elif (first == second_answer or first == third_answer or second == first_answer or second == third_answer or third == first_answer or third == second_answer):
        print("ball")
        life_count += 1 
    else:
        print("out")
        out_count += 1


    
```

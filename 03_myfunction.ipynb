잃어버린 보물 찾기

```python
import time

# 초기 체력
heart = 100

# 체력 감소 함수
def lose_health(amount):
    global heart
    heart -= amount
    print(f"⚠️ 체력이 {amount} 줄었습니다. 현재 체력: {heart}")
    if heart <= 0:
        print("💀 체력이 모두 소진되었습니다. 게임 오버!")
        exit()

# 게임 시작
def start_game():
    global heart
    print("🏝️ 보물 찾기 게임에 오신 걸 환영합니다!")
    time.sleep(2)
    print("앞에는 두 갈래 길이 있습니다.")
    choice = input("어느 길로 가시겠습니까? (왼쪽/오른쪽): ")

    if choice == "왼쪽":
        left_path()
    elif choice == "오른쪽":
        right_path()
    else:
        print("❓ 올바르지 않은 선택입니다. 게임 오버!")
        exit()

# 왼쪽 길
def left_path():
    print("\n🌲 왼쪽 길로 들어섰습니다. 갑자기 함정이 발 밑에서!")
    lose_health(30)
    time.sleep(2)
    print("앞에 오래된 다리가 있습니다.")
    choice = input("다리를 건너시겠습니까? (예/아니오): ")

    if choice == "예":
        print("😱 다리가 무너졌습니다!")
        lose_health(50)
        print("겨우 살아남았습니다. 조심하세요...")
        cave()
    elif choice == "아니오":
        print("현명한 선택입니다. 옆길로 돌아갑니다.")
        cave()
    else:
        print("❓ 이상한 행동을 했습니다. 맹수에게 당했습니다.")
        exit()

# 오른쪽 길
def right_path():
    print("\n🌴 오른쪽 길은 한적합니다. 하지만 뱀이 나타났습니다!")
    lose_health(20)
    time.sleep(2)
    print("앞에 샘물이 있습니다.")
    choice = input("물을 마시겠습니까? (예/아니오): ")

    if choice == "예":
        print("💧 체력이 회복되었습니다!")
        global heart
        heart += 30
        if heart > 100:
            heart = 100
        print(f"현재 체력: {heart}")
        cave()
    elif choice == "아니오":
        print("물을 지나칩니다. 계속 이동합니다.")
        cave()
    else:
        print("❓ 망설이는 사이에 맹수에게 당했습니다.")
        exit()

# 동굴 진입
def cave():
    print("\n🕳️ 깊은 동굴을 발견했습니다.")
    choice = input("들어가시겠습니까? (예/아니오): ")

    if choice == "예":
        print("🗺️ 보물지도를 발견했습니다!")
        final_stage()
    elif choice == "아니오":
        print("동굴을 지나쳤지만, 갑자기 낙석이!")
        lose_health(50)
        final_stage()
    else:
        print("❓ 이상한 행동을 했습니다. 돌무더기에 깔렸습니다.")
        exit()

# 마지막 보물 선택
def final_stage():
    print("\n💎 세 개의 상자가 있습니다. 하나를 선택하세요.")
    choice = input("어느 상자를 여시겠습니까? (1/2/3): ")

    if choice == "2":
        print("🎉 축하합니다! 잃어버린 보물을 찾았습니다!")
        print("게임을 클리어하셨습니다! 👑")
    elif choice == '1':
        print("💣 꽝입니다! 다음 기회에")
    else:
        print("💣 함정이었습니다. 게임 오버!")
        return

# 게임 실행
start_game()
```

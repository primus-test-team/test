```python
def run_balance_mbti_test():
    print("🎮 밸런스 게임 스타일 MBTI 테스트 시작!\n")
    scores = {"E": 0, "I": 0, "S": 0, "N": 0, "T": 0, "F": 0, "J": 0, "P": 0}

    questions = [
        {
            "question": "1. 친구들과 밤새 수다 떨기 vs 혼자 조용히 영화 보기",
            "A": "E",
            "B": "I"
        },
        {
            "question": "2. 확실한 사실에 기반해 판단하기 vs 직관적으로 감 잡기",
            "A": "S",
            "B": "N"
        },
        {
            "question": "3. 논리적인 해석이 우선! vs 감정과 상황을 고려해야지",
            "A": "T",
            "B": "F"
        },
        {
            "question": "4. 계획 세우고 움직이기 vs 일단 해보고 생각하기",
            "A": "J",
            "B": "P"
        },
        {
            "question": "5. 모르는 사람과도 쉽게 대화 가능 vs 아는 사람이 더 편해",
            "A": "E",
            "B": "I"
        },
        {
            "question": "6. 구체적인 설명이 좋아 vs 대충 감으로 알아들음",
            "A": "S",
            "B": "N"
        },
        {
            "question": "7. 사실과 원칙이 중요함 vs 사람 마음이 더 중요하지",
            "A": "T",
            "B": "F"
        },
        {
            "question": "8. 미리 준비하는 게 편해 vs 즉흥이 더 재밌지",
            "A": "J",
            "B": "P"
        },
    ]

    for q in questions:
        while True:
            print(q["question"])
            answer = input("👉 A or B 중 선택: ").upper()
            if answer == "A":
                scores[q["A"]] += 1
                break
            elif answer == "B":
                scores[q["B"]] += 1
                break
            else:
                print("A 또는 B를 입력해주세요!\n")

    # MBTI 계산
    mbti = ""
    mbti += "E" if scores["E"] >= scores["I"] else "I"
    mbti += "S" if scores["S"] >= scores["N"] else "N"
    mbti += "T" if scores["T"] >= scores["F"] else "F"
    mbti += "J" if scores["J"] >= scores["P"] else "P"

    print(f"\n✅ 당신의 MBTI는 {mbti} 입니다!")

run_balance_mbti_test()
```
